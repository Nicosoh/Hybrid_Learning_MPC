```python
import matplotlib.pyplot as plt

def latex_to_image(latex_str, save_path="equation.svg", dpi=300):
    # Use matplotlib's built-in engine instead of a local system LaTeX install
    plt.rcParams['mathtext.fontset'] = 'cm'  # 'cm' = Computer Modern, the classic LaTeX font
    
    # Create an ultra-tiny canvas that expands to fit the text
    fig, ax = plt.subplots(figsize=(0.01, 0.01)) 
    
    # Render the raw string using standard matplotlib math parsing
    ax.text(0, 0, f"${latex_str}$", fontsize=24, ha='center', va='center')
    
    # Strip the borders and backgrounds
    ax.axis('off')
    
    # Save as high-res PNG or infinite-scale SVG
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0.05, transparent=True, dpi=dpi)
    plt.close()

# Example usage (Note the raw 'r' string to handle backslashes correctly)
latex_to_image(r"J_\infty(x_0) = \lim_{H \to \infty} \; \min_{u_0, \dots, u_{H-1}} \sum_{t=0}^{H-1} l(x_t, u_t) + l_H(x_H)", "variance.png")

# \begin{equation}
# V_{\theta}(x_{t}) = l(x_t, u_t) + V_{\theta^-}(x_{t+1})
# \end{equation}

# \begin{equation}
# \theta^- \leftarrow (1 - \tau)\theta^- + \tau \theta
# \end{equation}
```
